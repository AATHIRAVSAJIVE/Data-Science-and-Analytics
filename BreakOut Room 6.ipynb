```python
sns.pairplot(data)
plt.show()
```

```python
plt.bar(data['day'],data['tip'],color = 'purple')

plt.title('BAR CHART OF TIP BY DAY',size = 15,color ='g')
plt.xlabel('DAY')
plt.ylabel('TIP')

plt.show() 
```

```python
plt.bar(data['size'],data['total_bill'])

plt.title('BAR CHART OF TOTAL BILL BY SIZE',size = 15,color ='g')
plt.xlabel('SIZE')
plt.ylabel('TOTAL BILL')

plt.show() 
```

```python
plt.figure(figsize=(8,4))

plt.hist(data['total_bill'],color = 'g',rwidth =0.98)

plt.title('HISTOGRAM OF TOTAL BILL',size = 15,color ='b')
plt.xlabel('TOTAL BILL')
plt.ylabel('COUNT')

plt.show() 
```

```python
plt.figure(figsize=(8,4))

plt.hist(data['tip'],color = 'orange',rwidth =0.98)

plt.title('HISTOGRAM OF TIP',size = 15,color ='b')
plt.xlabel('TIP')
plt.ylabel('COUNT')

plt.show() 
```

```python
plt.figure(figsize = (8,4))

sns.countplot(x ="sex",data = data,width = 0.5)

plt.title('COUNT PLOT OF BILL PAYER BY SEX',size = 15,color ='r')

plt.show()
```

```python
plt.figure(figsize = (8,4))

sns.countplot(x ="smoker",data = data,width = 0.5)

plt.title('COUNT PLOT OF SMOKERS PRESENT',size = 15,color ='r')

plt.show()
```

```python
plt.figure(figsize = (8,4))

sns.countplot(x ="time",data = data,width = 0.5)

plt.title('COUNT PLOT OF BILL PAYER BY TIME',size = 15,color ='r')

plt.show()
```

```python
sns.barplot(x = 'size', y = 'tip',data = data,width = 0.5)

plt.title('PLOT OF TIP BY SIZE',size = 15,color ='purple')

plt.show()
```

```python
sns.scatterplot(x="tip", y="total_bill", data=data)

plt.title('SCATTER PLOT OF TOTAL BILL BY TIP',size = 15,color ='purple')

plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=data)

plt.title('BOX PLOT OF TOTAL BILL  BY DAY',size = 15,color ='purple')

plt.show()
```
