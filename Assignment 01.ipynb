```python
# Q-1. Read the dataset to python environment.
```

```python
# first step import libraries 
    
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns          

# to read the dataset to python environment,use pandas library of python

data = pd.read_excel(r'C:\Users\sajee\Downloads\AATHIRA DSA\iris.xls')     # storing the result in the variable 'data'

print(data)
```

```python
# Q-2. Display the columns in the dataset.
```

```python
# to display the columns in the dataset

# in pandas library '.columns' is used to display the column labels of a DataFrame

column_data = data.columns     # storing the result in the variable 'column_data'

print(column_data)
```

```python
# Q-3. Calculate the mean of each column of the dataset.
```

```python
# to calculate the mean of each column of the dataset

 # '.mean()' is a method in Python that calculate the arithmetic mean of each column of the DataFrame

mean_data = data.mean(numeric_only=True)     # storing the result in the variable 'mean_data'    

print(mean_data)
```

```python
# explanation for above problem 

data.info()     # in pandas library '.info()' provides summary of DataFrame

# output shows that 4 th column index is object,
# when calculating the mean using 'mean' in pandas, columns that contain object (string) data will be ignored by default
```

```python
# Q-4. Check for the null values present in the dataset.
```

```python
# to check for the null values present in the dataset

# '.isna()' is a method in pandas used to check for missing or null values
# '.sum()' is a method in pandas used to calculate the sum of the elements
# '.isna().sum()' is a method used to count the number of missing or null values in each column of a DataFrame

nullvalues_data = data.isna().sum()     # storing the result in the variable 'nullvalues_data' 

print(nullvalues_data)

#Since there are no null values in the DataFrame,the '.isna().sum()' returns zero
```

```python
# Q-5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.
```

```python
# 5:1     for visualizations of dataset , use seaborn data visualization library of python

# 'sns.barplot()' is a function in the seaborn library of Python used to create bar plot

# to create a bar plot of petal length by species using seaborn pass the code for x-axis and y-axis

sns.barplot(x = 'Classification', y = 'PL',data = data,width = 0.4)    # width parameter used to set the width of bar

# set the title,xlabel and ylabel of the plot using Matplotlib's title(), xlabel(), and ylabel() functions, respectively

plt.title('PETAL LENGTH BY SPECIES',size=15,c='red')     # setting size and color of title along with it
plt.xlabel('CLASSIFICATION OF SPECIES')
plt.ylabel('PETAL LENGTH MEASUREMENTS')

plt.show()     # to show the plot
```

```python
# 5:2    for visualizations of dataset , use seaborn data visualization library of python

# create a box plot using the 'sns.boxplot()' function available seaborn library of Python used to create box plot

# creating a box plot of iris data set using seaborn library 

sns.boxplot(data)

# set the title,xlabel and ylabel of the plot using Matplotlib's title(), xlabel(), and ylabel() functions, respectively

plt.title('CLASSIFICATION OF IRIS DATA SET',size=15,c='red')     # setting size and color of title along with it
plt.xlabel('CLASSIFICATION')
plt.ylabel('MEASUREMENTS')


plt.show()     # to show the plot
```

```python
# 5:3     for visualizations of dataset , use matplotlib data visualization library of python

# 'plt.hist()' is a function in the matplotlib library of Python used to create histograms

# creating a histogram for sepal length from the iris dataset

# 'rwidth' parameter to set the width of bar and 'color' parameter to set color

plt.hist(data['SL'] , color = 'g',rwidth =0.97)      

# set the title,xlabel and ylabel of the plot using Matplotlib's title(), xlabel(), and ylabel() functions, respectively

plt.title('IRIS DATA OF SEPAL LENGTH',size = 15,color ='r')      # setting size and color of title along with it
plt.xlabel('SEPAL LENGTH')
plt.ylabel('COUNT')

plt.show()     # to show the plot
```

```python
# 5:4     for visualizations of dataset , use matplotlib data visualization library of python

# 'plt.scatter' is a function in the matplotlib library of Python used to create scatter plots

#  create a scatter plot of SL and SW & create a scatter plot of PL and PW

# 's' parameter to set size  and 'marker' parameter to set marker
# set label and color of plot using 'label' and 'color' parameters

plt.scatter(data['PL'], data['PW'],s=25,marker = "*", label='Petal', color='green')

plt.scatter(data['SL'], data['SW'],s=25,marker = "x" ,label='Sepal', color='blue')

# set the title,xlabel and ylabel of the plot using Matplotlib's title(), xlabel(), and ylabel() functions, respectively

plt.title('IRIS DATA MEASUREMENTS',size = 15,color ='r')     # setting size and color of title along with it
plt.xlabel('LENGTH')
plt.ylabel('WIDTH')

plt.legend()     # legend method is used to display legend for the two scatter plots.

plt.show()     # to show the plot
```

```python
# 5:5     for visualizations of dataset , use seaborn data visualization library of python

# 'sns.countplot' is a function from Seaborn is used to plot the count of occurrences of categorical data in a dataset

#  to plot the count of occurrences of the 'Classification'column in a DataFrame called 'data' pass the code for x-axis

sns.countplot(x = 'Classification',data = data)     # plot a bar chart of the counts

# set the title of the plot using Matplotlib's 'title()' function , set the size and color of title along with it

plt.title('COUNT OF OCCURRENCES OF CLASSIFICATION',size = 15,color ='r')     

plt.show()     # to show the plot
```
